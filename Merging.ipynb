```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading all the input files
data1 = pd.read_csv('Dataset/assessments.csv')
data2 = pd.read_csv('Dataset/courses.csv')
data3 = pd.read_csv('Dataset/studentAssessment.csv')
data4 = pd.read_csv('Dataset/studentInfo.csv')
data5 = pd.read_csv('Dataset/studentRegistration.csv')
data6 = pd.read_csv('Dataset/studentVle.csv')
data7 = pd.read_csv('Dataset/vle.csv')
```

```python
#shape of all the files
#shape of all the files
frames = [data1,data2,data3,data4,data5,data6,data7]
for i in frames:
    print(i.shape)
```

```python
#merging the datasets assessment and student assessment
df = pd.merge(data1, data3,
                   on='id_assessment', 
                   how='inner')
```

```python
#renaming the coulmn date to avoind redundancy
df.rename(columns = {'date': 'date_assessment'}, inplace = True)
```

```python
#grouping student vle data based on the given attributes
df1 = data6.groupby(['code_module','code_presentation','id_student','date'])['sum_click'].sum().reset_index()
```

```python
#merging student vle data and assessment data
df3 = pd.merge(df1,df,on=['id_student','code_module','code_presentation'],how='inner')
```
